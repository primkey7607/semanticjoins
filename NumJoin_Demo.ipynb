## Semantic Labeling
NumJoin chooses the most likely semantic label (DBpedia class property label) for each column of every candidate table (the likelihood is called a 'membership probability'). Semantic labels and likelihoods for an example table are shown below.

```python
from numjoin_apis import show_semantic_labels
show_semantic_labels('demo_lake/busridertbl.csv')
```

Then, for each candidate table, NumJoin computes the entropies of the semantic label likelihoods for each column, and averages them across candidate table columns. If this average entropy is below some chosen entropy threshold, NumJoin uses the KG-based approach. Otherwise, NumJoin uses the Distribution-based approach.

```python
from numjoin_apis import avg_member_entropy

avg_member_entropy('demo_lake/busridertbl.csv', 
                   'demo_lake/busriderjoin.csv')
```

## KG-based Approach: Numeric Disambiguation
If NumJoin chooses the KG-based approach for a candidate table, then for each row in the join, NumJoin will use the semantic labels and cell values to disambiguate the row to at least two real-world entities, (at least one per table involved in the join). In the example below, the first row of the join between tables busridertbl.csv and busriderjoin.csv is disambiguated as bus company 'Shuttle-UM', and location 'Montgomery County, Maryland'. 

```python
import pandas as pd
from numjoin_apis import disambiguate_onerow

indf = pd.read_csv('demo_lake/busridertbl.csv')
outdf = pd.read_csv('demo_lake/busriderjoin.csv')
joindf = indf.merge(outdf, left_on='dbo:regionServed', right_on='dbo:regionServed')
row0 = joindf.to_dict(orient='records')[0]

disambiguate_onerow('demo_lake/busridertbl.csv', 'demo_lake/busriderjoin.csv', row0)
```

## KG-based Approach: Instance Correspondence
Once NumJoin has found at least one pair of real-world entities (instances) for each row of the join, it will find relationships for each pair of entities. For example, between entities 'Shuttle-UM' and 'Montgomery County, Maryland', NumJoin finds two possible relationships: 'regionServed', and 'serviceArea'.

```python
from numjoin_apis import find_relationships

find_relationships({('dbo:BusCompany',): ['http://dbpedia.org/resource/Shuttle–UM']},
                  {(): ['http://dbpedia.org/resource/Montgomery_County,_Maryland']})
```

## KG-based Approach: KG Scoring
NumJoin scores the join based on the highest proportion of records in the join with the same relationship. In this case, the 'serviceArea' and 'regionServed' relationships were found in 184 out of 202 records, so the KG score is 184 / 202 = 0.91, roughly.

```python
from numjoin_apis import find_kgstrength
find_kgstrength('demo_lake/busridertbl.csv', 'demo_lake/busriderjoin.csv')
```

## KG-based Approach: Example Result
We provide the APIs above to run various parts of NumJoin's KG-based approach. Given the input table and a candidate, one can run the full KG-based approach using the below API. This visualizes the entities NumJoin finds in the green column. The join keys are shown in yellow.

```python
from numjoin_apis import show_kgmatch
show_kgmatch('demo_lake/busridertbl.csv', 'demo_lake/busriderjoin.csv')
```

## KG-based Approach Full Pipeline: Finding the Top-k
Given an input table and number of join candidates to return, k, NumJoin will return up to k joins with the highest scores. Below, we see that given input table 'demo_lake/busridertbl.csv' and k = 3, NumJoin finds only one table 'demo_lake/busriderjoin.csv' with score 0.91.

```python
from numjoin_apis import show_topk_kgmatches
show_topk_kgmatches('demo_lake/busridertbl.csv', 3)
```

## Distribution-based Approach: Proxy Finding
Given an input table 'demo_lake/ETF prices.csv', NumJoin first finds a proxy table that is most similar to the input based on numeric types and distributions of column values. The distributions are described using numeric features (e.g., number of unique values and their frequencies), and NumJoin finds the normalized difference between numeric features of the input table, and those of the join candidate. In this case, the join candidate is 'demo_lake/IndexProcessed.csv', and the normalized difference is 0.112529.

```python
from numjoin_apis import find_proxy_table
find_proxy_table('demo_lake/ETF prices.csv')
```

## Distribution-based Approach: Proxy-Input Joins
Given the proxy table, NumJoin then finds Minhash LSH-based join candidates. Similar to the above, we show the join candidate ('Output table'), its join key, ('Output Table Join Key'), and the input table join key for the candidate ('Input Table Join Key').

```python
show_lshjoins('demo_lake/indexProcessed.csv')
```

## Distribution-based Approach: Distribution Scoring
NumJoin scores the join based on the join cardinality, which is the number of records in the join divided by the product of the numbers of records in both the input and candidate tables. For example, the score of the join between 'demo_lake/ETF prices.csv' and 'demo_lake/MutualFund prices - A-E.csv' is roughly 0.0001, as shown below.

```python
from numjoin_apis import find_diststrength
find_diststrength('demo_lake/ETF prices.csv', 'demo_lake/MutualFund prices - A-E.csv')
```

## Distribution-based Approach: Example Result
We provide the APIs above to run various parts of NumJoin's distribution-based approach. Given the input table and a candidate, one can run the full KG-based approach using the below API. This visualizes the proxy table and join table NumJoin finds. The join keys are shown in yellow.

```python
from numjoin_apis import show_distmatch
show_distmatch('demo_lake/ETF prices.csv', 'demo_lake/MutualFund prices - A-E.csv')
```

```python
from numjoin_apis import show_topk_distmatches
show_topk_distmatches('demo_lake/ETF prices.csv', 3)
```
